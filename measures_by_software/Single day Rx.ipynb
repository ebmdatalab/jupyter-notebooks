```python
##importing libraries
import pandas as pd
import numpy as np
from ebmdatalab import bq, maps, charts
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
sql = '''
SELECT
  PCO_Name,
  PCO_Code,
  Practice_Name, 
  Practice_Code,
  presc.bnf_code,
  dmdone.id,
  dmdtwo.hosp,
  bnf_description AS bnf_name,
  Items,
  Quantity
FROM
  ebmdatalab.tmp_eu.raw_prescribing_data_2019_04 AS presc
Left JOIN  
  ebmdatalab.dmd.ampp AS dmdone
ON
presc.bnf_code = dmdone.bnf_code
LEFT JOIN
ebmdatalab.dmd.prescribinfo AS dmdtwo
ON
dmdone.id = dmdtwo.ampp 
WHERE
presc.Quantity <= 1 AND dmdtwo.hosp = true
'''

df1 = bq.cached_read(sql, csv_path='singledayrx.csv')
df1.rename(columns={'PCO_Code':'pct'}, inplace=True)
df1.head(5)
```
