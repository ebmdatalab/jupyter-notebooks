```python
##importing libraries
import pandas as pd
import numpy as np
from ebmdatalab import bq, maps, charts
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
sql = '''
SELECT
pct_id, 
month, 
numerator, 
denominator, 
calc_value,
percentile
FROM
`ebmdatalab.measures.ccg_data_libre`
'''

df_libre = bq.cached_read(sql, csv_path='libre_latestmonth.csv')
df_libre.rename(columns={'pct_id':'pct'}, inplace=True)
df_libre.head()
```

```python
plt.figure(figsize=(12, 7))
last_month = df_libre.loc[df_libre['month']==df_libre['month'].max()]
maps.ccg_map(last_month, title="April 2019: Libre Measure",
             column='calc_value', 
             separate_london=True)

plt.show()
```
